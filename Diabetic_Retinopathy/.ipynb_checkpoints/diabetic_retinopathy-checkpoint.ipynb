# Diabetic Retinopathy
Retinopathy is an eye condition that causes changes to the blood vessels in the part of the eye called the retina.
This often leads to vision changes or blindness.
Diabetic patients are known to be at high risk for retinopathy. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
DIR = './data/'

X = pd.read_csv(DIR + 'x.csv', index_col=[0])
y = pd.read_csv(DIR +'y.csv', index_col=[0])
y = y['y']
```

For y:

1 - The patient has diabetic retinopathy

0 - The patient does not have diabetic retinopathy

Splitting our data into train and test sets, making sure that the class distribution are stratified

```python
train_x, test_x, train_y, test_y = train_test_split(X, y, test_size=0.25, random_state=0, stratify=y)
```

```python
train_y.value_counts()
```

Plotting the histograms of each column of train_x below

```python
for col in train_x.columns:
    train_x.loc[:, col].hist()
    plt.title(col)
    plt.show()
```

As we can see, the distributions have a generally bell shaped distribution, but with slight rightward skew.
We would transform the data using log transformation

```python
train_x = np.log(train_x)
test_x = np.log(test_x)

mean_ = train_x.mean()
std_ = train_x.std()
```

```python
train_x = (train_x - mean_) / std_
test_x = (test_x - mean_) / std_
```

Replotting the distributions of the train data to confirm normality

```python
for col in train_x.columns:
    train_x.loc[:, col].hist()
    plt.title(col)
    plt.show()
```

### Base Model

For base model, we'll go with logistic regression since our task is binary classification

```python
model = LogisticRegression(random_state=2020)
model.fit(train_x, train_y)
```

### Model Evaluation using the C-index

Now that we have a model, we need to evaluate it. We'll do this using the **c-index.** 
* The c-index measures the discriminatory power of a risk score. 
* Intuitively, a higher c-index indicates that the model's prediction is in agreement with the actual outcomes of a pair of patients.
* The formula for the c-index is

$$ \mbox{cindex} = \frac{\mbox{concordant} + 0.5 \times \mbox{ties}}{\mbox{permissible}} $$

* A permissible pair is a pair of patients who have different outcomes.
* A concordant pair is a permissible pair in which the patient with the higher risk score also has the worse outcome.
* A tie is a permissible pair where the patients have the same risk score.


```python
def cindex(y_true, scores):
    n = len(y_true)
    assert len(scores) == n

    concordant = 0
    permissible = 0
    ties = 0
    
    for i in range(n):
        for j in range(i+1, n):
            if y_true[j] != y_true[i]:
                permissible += 1
                if scores[j] == scores[i]:
                    ties += 1
                    continue
                if y_true[i] == 0 and y_true[j] == 1:
                    if scores[i] < scores[j]:
                        concordant += 1
                if y_true[i] == 1 and y_true[j] == 0:
                    if scores[i] > scores[j]:
                        concordant += 1
                        
    c_index = (concordant + (0.5 * ties)) / permissible
    
    return c_index
```

```python
scores = model.predict_proba(test_x)[:, 1]
c_index_test_x = cindex(test_y.values, scores)
print(f"c-index on test set is {c_index_test_x:.4f}")
```

Plot the coefficients to see which features are having the most effects

```python
coeffs = pd.DataFrame(data = model.coef_, columns = train_x.columns)
coeffs.T.plot.bar(legend=None);
```

### Improving the model performance using Feature Engineering

**We start by testing the effect of interactions**

```python
def add_interactions(X):
    features = X.columns
    m = len(features)
    X_int = X.copy(deep=True)

    # 'i' loops through all features in the original dataframe X
    for i in range(m):
        feature_i_name = features[i]
        feature_i_data = X[feature_i_name]
        for j in range(i+1, m):
            feature_j_name = features[j]
            feature_j_data = X[feature_j_name]
            feature_i_j_name = f"{feature_i_name}_x_{feature_j_name}"
            X_int[feature_i_j_name] = feature_i_data * feature_j_data

    return X_int
```

```python
print("Original Data")
print(train_x.loc[:, ['Age', 'Systolic_BP']].head(), "\n")
print("Data with Interactions")
print(add_interactions(train_x.loc[:, ['Age', 'Systolic_BP']].head()))
```

```python
X_train_int = add_interactions(train_x)
X_test_int = add_interactions(test_x)
```

```python
model_X_int = LogisticRegression(random_state=2020).fit(X_train_int, train_y)
```

```python
scores_X = model.predict_proba(test_x)[:, 1]
c_index_X_test = cindex(test_y.values, scores_X)

scores_X_int = model_X_int.predict_proba(X_test_int)[:, 1]
c_index_X_int_test = cindex(test_y.values, scores_X_int)

print(f"c-index on test set without interactions is {c_index_X_test:.4f}")
print(f"c-index on test set with interactions is {c_index_X_int_test:.4f}")
```

We can see that interactions improved the model, so we would add them.

```python
train_x = X_train_int
test_x = X_test_int

model = model_X_int
```

Now let us check which features helped the model

```python
coeffs = pd.DataFrame(data = model.coef_, columns = train_x.columns)
coeffs.T.plot.bar(legend=None);
```

**Now we test the effect of age binning using IQR**

We'll generate features such as mean encodings with the age binning on important features shown above

```python
q1 = train_x['Age'].quantile(.25)
q2 = train_x['Age'].quantile(.50)
q3 = train_x['Age'].quantile(.75)

def age_group(x):
    if (x < q1):
        return "group1"
    if (x >= q1 and x < q2):
        return "group2"
    if (x >= q2 and x < q3):
        return "group3"
    if (x > q3):
        return "group4"
    
train_x['Age_group'] = train_x['Age'].apply(age_group)
test_x['Age_group'] = test_x['Age'].apply(age_group)
train_x.head()
```

```python
age_mean_systolic_bp = train_x.groupby('Age_group')['Systolic_BP'].mean()
age_mean_cholesterol = train_x.groupby('Age_group')['Cholesterol'].mean()
age_mean_age_x_cholesterol = train_x.groupby('Age_group')['Age_x_Cholesterol'].mean()
```

```python
train_x['age_mean_systolic_bp'] = train_x['Age_group'].map(age_mean_systolic_bp)
train_x['age_mean_cholesterol'] = train_x['Age_group'].map(age_mean_cholesterol)
train_x['age_mean_age_x_cholesterol'] = train_x['Age_group'].map(age_mean_age_x_cholesterol)

test_x['age_mean_systolic_bp'] = test_x['Age_group'].map(age_mean_systolic_bp)
test_x['age_mean_cholesterol'] = test_x['Age_group'].map(age_mean_cholesterol)
test_x['age_mean_age_x_cholesterol'] = test_x['Age_group'].map(age_mean_age_x_cholesterol)
```

```python
clf = LogisticRegression(random_state=2020)
train_x = train_x[[col for col in train_x.columns if not col == "Age_group"]]

clf.fit(train_x, train_y)
```

```python
test_x = test_x[[col for col in train_x.columns if not col == "Age_group"]]

scores_X = clf.predict_proba(test_x)[:, 1]
c_index_X_test = cindex(test_y.values, scores_X)

print(f"c-index on test set without interactions is {c_index_X_test:.4f}")
```

We can see that binning helps it improve based on c-index so we add it

```python
model = clf
```

```python
coeffs = pd.DataFrame(data = model.coef_, columns = train_x.columns)
coeffs.T.plot.bar(legend=None);
```
